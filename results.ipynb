# Stable Diffusion Results

```python
# Using readlines()
file1 = open('sd_output_log.txt', 'r')
Lines = file1.readlines()
  
# Strips the newline character
male_dict = {}
female_dict = {}
gender = ''
for line in Lines:
    if '_man' in line or '_boy' in line or '_men' in line or '_male' in line:
        gender = 'm'
    elif '_woman' in line or '_girl' in line or '_women' in line or '_female' in line:
        gender = 'f'
    elif gender == 'm':
        x = line.split(":")
        key = x[0]
        if key in male_dict:
            male_dict[key] += 1
        else:
            male_dict[key] = 1
    elif gender == 'f':
        x = line.split(":")
        key = x[0]
        if key in female_dict:
            female_dict[key] += 1
        else:
            female_dict[key] = 1
    
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
plt.bar(range(len(male_dict)), list(male_dict.values()), align='center')
plt.xticks(range(len(male_dict)), list(male_dict.keys()))
# # for python 2.x:
# plt.bar(range(len(D)), D.values(), align='center')  # python 2.x
# plt.xticks(range(len(D)), D.keys())  # in python 2.x

plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.bar(range(len(female_dict)), list(female_dict.values()), align='center')
plt.xticks(range(len(female_dict)), list(female_dict.keys()))
# # for python 2.x:
# plt.bar(range(len(D)), D.values(), align='center')  # python 2.x
# plt.xticks(range(len(D)), D.keys())  # in python 2.x

plt.show()
```

```python
all_keys = []
for k in female_dict:
    all_keys.append(k)
for k in male_dict:
    if k not in all_keys:
        all_keys.append(k)
```

```python
import numpy as np
male_biases = {}
female_biases = {}
for k in all_keys:
    m = 0
    f = 0
    if k == 'person':
        continue

    if k in female_dict :
        f = female_dict[k]
    if k in male_dict:
        m = male_dict[k]

    male_bias = (m - f) / (m + f) 
    if (m + f) > 4:
        male_biases[k] = male_bias
        female_biases[k] = 0
    if (m + f) > 4:
        male_biases[k] = 0
        female_biases[k] = male_bias
```

```python
f_bar =  {} 
m_bar = {} 
for k in all_keys:
    if k in female_biases and k != 'person':
        try:
            f_bar[k] = female_dict[k]
        except:
            f_bar[k] = 0


for k in all_keys:
    if k in male_biases and k != 'person':
        try:
            m_bar[k] = male_dict[k]
        except:
            m_bar[k] = 0
```

```python
plt.figure(figsize=(14,6))
plt.bar(np.arange(len(m_bar))-0.2, list(m_bar.values()), width=0.4, align='center', color='#10559A', label="Male Prompts")
plt.xticks(np.arange(len(m_bar)), list(m_bar.keys()))
plt.bar(np.arange(len(f_bar))+0.2, list(f_bar.values()), width=0.4, align='center', color='#DB4C77', label="Female Prompts")

plt.legend(loc="upper right")
plt.axhline(y=0,linewidth=1, color='k', linestyle='--')
plt.xticks(rotation = 90)
plt.xlabel("Object Categories")#, fontsize=12)
plt.ylabel("Detection Count")#, fontsize=12)
plt.title("Objects detected in Stable Diffusion images",fontdict={'weight': 'bold'})#, fontsize=18)
plt.show()
```
